### Quadrotor Coordinate Frames

I use the NED convention and define the following frames:<br>
1. The intertial frame $F^{i}$
2. The vehicle frame
3. the vehicle-1 frame
4. The vehicle-2 frame
5. The body frame

Reference : [link](https://scholarsarchive.byu.edu/cgi/viewcontent.cgi?article=2324&context=facpub)
Remember: psi represents yaw, theta represents pitch, phi represents roll

#### For now we assume inertial frame to be same as vehicle frame

```python
def vehicle_to_vehicle1_transformation(psi):
     return np.array(
                [    [cos(psi) , sin(psi), 0],
                     [-sin(psi), cos(psi), 0],
                     [0      , 0     , 1]  ]  )
    
def vehicle1_to_vehicle2_trtansformation(theta):
     return np.array(
                    [[cos(theta) , 0, -sin(theta)],
                    [0, 1, 0],
                    [sin(theta)    , 0   , cos(theta)]]  )
    
def vehicle2_to_body_transformation(phi):
     return np.array(
                    [[1 , 0, 0],
                    [0, cos(phi), sin(phi)],
                    [0 , -sin(phi), cos(phi)]]  )
    
def vehcile_to_body_transformation(phi,theta,psi):
    return np.dot(vehicle2_to_body_transformation(phi), 
                  vehicle1_to_vehicle2_trtansformation(theta),
                  vehicle_to_vehicle1_transformation(psi)  )
    
```
